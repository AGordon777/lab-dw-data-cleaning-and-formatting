## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they contain only valid data. For example, the gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized.

```python
# Your code here
```

```python
customer.isna().sum()
#lots of missing data - I'm going to remove lines where the Customer field is blank
```

```python
#delete the empty lines to better see the data cleanliness, then check
customer = customer[customer['Customer'].isna()==False]
customer.isna().sum()
```

```python
#now we are going to clean up our category data, starting with Gender
customer['Gender'].unique()
```

```python
#using a cleaning function to tidy up the Gender elements

def clean_gender(x):
    if x in ['M', 'MALE','Male']:
        return 'M'
    elif x in ['Femal','female','F']:
        return 'F'
    else:
        return x

customer['Gender'] = list(map(clean_gender, customer['Gender']))
customer['Gender'].unique()
```

```python
customer['State'].unique()
```

```python
#using a cleaning function to tidy up the State elements

def clean_state(x):
    if x in ['AZ']:
        return 'Arizona'
    elif x in ['WA']:
        return 'Washington'
    elif x in ['Cali']:
        return 'California'
    else:
        return x

customer['State'] = list(map(clean_state, customer['State']))
customer['State'].unique()
```

```python
#checking the others - can see Bachelor and Bachelors in Education
display(customer['Education'].unique())
display(customer['Policy Type'].unique())
display(customer['Vehicle Class'].unique())
```

```python
#replace the erroneous Education category
customer['Education'].replace('Bachelors','Bachelor',inplace=True)
customer['Education'].unique()
```

# Challenge 2: creating functions on a separate `py` file

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting. 

Write these functions in a separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain.

```python
# Your code here

#below are all the functions done to the data above as comments
#with associated functions beneath:

# import pandas as pd
# customer = pd.read_csv('customer_data.csv')
    
# customer = customer.rename(columns={'ST':'State','GENDER':'Gender'})

# customer = customer[customer['Customer'].isna()==False]

# customer['Gender'] = list(map(clean_gender, customer['Gender']))

# customer['State'] = list(map(clean_state, customer['State']))

# customer['Education'].replace('Bachelors','Bachelor',inplace=True)

# customer['Number of Open Complaints'].replace('1/0/00','0',inplace=True)
# customer['Number of Open Complaints'].replace('1/1/00','1',inplace=True)
# customer['Number of Open Complaints'].replace('1/2/00','2',inplace=True)
# customer['Number of Open Complaints'].replace('1/3/00','3',inplace=True)
# customer['Number of Open Complaints'].replace('1/4/00','4',inplace=True)
# customer['Number of Open Complaints'].replace('1/5/00','5',inplace=True)
# customer['Number of Open Complaints'] = pd.to_numeric(customer['Number of Open Complaints'], errors='coerce')

# customer['Customer Lifetime Value'] = customer['Customer Lifetime Value'].str.rstrip('%').astype('float') / 100.0

# customer['Gender'] = customer['Gender'].fillna('Unknown')

# customer['Customer Lifetime Value'] = customer['Customer Lifetime Value'].fillna('0')

# customer = customer.drop_duplicates()

def clean_gender(x):
    if x in ['M', 'MALE','Male']:
        return 'M'
    elif x in ['Femal','female','F']:
        return 'F'
    else:
        return x
    
def clean_state(x):
    if x in ['AZ']:
        return 'Arizona'
    elif x in ['WA']:
        return 'Washington'
    elif x in ['Cali']:
        return 'California'
    else:
        return x


def clean_data(x):
    x = x.rename(columns={'ST': 'State', 'GENDER': 'Gender'})
    x = x[x['Customer'].isna()==False]
    x['Education'].replace('Bachelors','Bachelor',inplace=True)
    x['Gender'] = list(map(clean_gender, x['Gender'])) #function is defined earlier
    x['State'] = list(map(clean_state, x['State'])) #function is defined earlier
    x['Gender'] = x['Gender'].fillna('Unknown')
    x['Number of Open Complaints'].replace('1/0/00','0',inplace=True)
    x['Number of Open Complaints'].replace('1/1/00','1',inplace=True)
    x['Number of Open Complaints'].replace('1/2/00','2',inplace=True)
    x['Number of Open Complaints'].replace('1/3/00','3',inplace=True)
    x['Number of Open Complaints'].replace('1/4/00','4',inplace=True)
    x['Number of Open Complaints'].replace('1/5/00','5',inplace=True)
    x['Number of Open Complaints'] = pd.to_numeric(x['Number of Open Complaints'], errors='coerce')
    x['Customer Lifetime Value'] = x['Customer Lifetime Value'].fillna('0')
    x['Customer Lifetime Value'] = x['Customer Lifetime Value'].str.rstrip('%')
    x['Customer Lifetime Value'] = pd.to_numeric(x['Customer Lifetime Value'], errors='coerce') / 100
    x = x.drop_duplicates()
    return x


# customer.to_csv('customer_clean.csv', index=False)
```
